```python
# Bibliotecas
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Configuração dos dados (N inicia em 1 para convergência no eixo)
n = np.array([1, 10, 100, 1000, 10000, 100000])

# Funções baseadas na sua tabela (ajustadas para precisão matemática)
complexidades = {
    'O(1)': np.ones_like(n),
    'O(log n)': np.log2(n) + 1,  # +1 para alinhar visualmente o início no ponto (1,1)
    'O(n)': n,
    'O(n log n)': n * (np.log2(n) + 1),
    'O(n^2)': n**2
}
```

```python
plt.figure(figsize=(10, 7), dpi=100)

# 2. Plotagem com estilo
for label, valores in complexidades.items():
    plt.plot(n, valores, label=label, marker='o', linewidth=2, markersize=5)

# 3. O Truque da "Junção dos Eixos"
plt.xscale('log')
plt.yscale('log')

# Definimos os limites exatamente onde os dados começam
plt.xlim(left=1, right=100000)
plt.ylim(bottom=1)

# 4. Estética de Apresentação
plt.title('Análise de Complexidade de Algoritmos (Big O)', fontsize=14, pad=15)
plt.xlabel('Tamanho da Entrada (N)', fontsize=12)
plt.ylabel('Operações / Tempo (Escala Log)', fontsize=12)
plt.grid(True, which="both", ls="--", alpha=0.6)
plt.legend(loc='upper left', fontsize=11)

# Ajuste fino para não cortar os rótulos
plt.tight_layout()

plt.show()
```
